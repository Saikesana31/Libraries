```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = fetch_openml('car',as_frame = True).frame
```

```python
data
```

```python
columns_to_encode = ['lug_boot','safety']

encoder = OrdinalEncoder(
    categories = [
        ['small','med','big'],
        ['low','med','high']
    ]
)

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

## OneHotEncoder

```python
df = fetch_openml('adult',as_frame = True).frame
```

```python
df
```

```python
df.occupation.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
encode = OneHotEncoder(handle_unknown='ignore',sparse_output = False)

encoded_values = encode.fit_transform(df[['occupation','race']])
new_columns = encode.get_feature_names_out(['occupation','race'])
```

```python
encoded_values
```

```python
new_columns
```

```python
df_encoded = pd.DataFrame(encoded_values,columns = new_columns,index  = df.index)

df_final = pd.concat(
    [df.drop(columns = ['occupation','race']),df_encoded], axis = 1
)
```
