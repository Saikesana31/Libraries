## Shuffle the order of tensors

```python
tensor = tf.constant([[10,7],
                     [1,2],
                     [8,9]])
tf.random.set_seed(42)
tensor_shuffle = tf.random.shuffle(tensor,seed = 42)
tensor_shuffle
```

```python
tensor_1 = tf.random.Generator.from_seed(42)
k  = tensor_1.normal(shape = (3,2))
k
```

```python
tf.subtract(tensor_m,10)
```

```python
# Matrix Multiplication
```

```python
tensor_m @ tensor_m
```

```python
tf.matmul(tensor_m,tensor_m)
```

```python
X = tf.constant([[1,2],
                [3,4],
                [5,6]])
Y = tf.constant([[7,8],
                [9,10],
                [11,12]])
X.shape,Y.shape
```

```python
y = tf.reshape(Y,shape = (2,3))
```

```python
#matrix multification
tf.matmul(X,y)
```

```python
Y,y,tf.transpose(Y)
```

```python
X @ tf.transpose(Y)
```

```python
tf.matmul(X,tf.transpose(Y))
```

```python
X,Y
```

```python
tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
tf.matmul(X,tf.transpose(Y))
```

```python
tf.tensordot(X,tf.transpose(Y),axes = 1)
```

```python
tf.abs(E)
```

```python
# creating a random tensor

F = tf.constant(np.random.randint(0,100,size=50))
F
```

```python
tf.reduce_min(F)
```

```python
tf.reduce_max(F)
```

```python
tf.reduce_mean(F)
```

```python
tf.reduce_sum(F)
```

```python
import tensorflow_probability as tfp
```

```python
tfp.stats.variance(F)
```

```python
tf.math.reduce_std(tf.cast(F,dtype = tf.float32))
```

```python
tf.math.reduce_variance(tf.cast(F,dtype = tf.float32))
```
