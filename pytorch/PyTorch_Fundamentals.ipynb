# Reshapping,stacking,squeezing and unsqueezing tensor

```python
k = torch.arange(1,10)
k,k.shape
```

```python
#Reshape
k_reshaped = k.reshape(1,9)
k_reshaped,k_reshaped.shape
```

```python
#view
z = k.view(1,9) #this is the view of k tensor
z
```

```python
z[:,0]=5 #if we change the value in the view then the value in original tensor also changes because it keeps same memory.
```

```python
k
```

```python
#stacking
k_stacked = torch.stack([k,k,k,k],dim=0)
k_stacked
```

```python
#Squeeze
k.shape,k_reshaped.shape
```

```python
k_squeezed = k_reshaped.squeeze()
k_squeezed,k_squeezed.shape
```

```python
l = k.squeeze()
l,l.shape
```

```python
#unsqueeze
m = k_reshaped.unsqueeze(dim=0)
m,m.shape
```

```python
#permute
a = torch.rand(size = (224,224,3))
a_permuted = a.permute(2,1,0)
a.shape,a_permuted.shape
```
